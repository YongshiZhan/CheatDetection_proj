```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
def sim_score_add_test(tot_q, share_q, same_q, weight):
#    count = 0
    score_sum = 0
    for i in range(same_q):
        score_sum += (1 + weight * i)
    return score_sum / (share_q + 1)
    
```

```python
weight = 0.2

all_score = []
tot_q = 20
x = []
for i in range(tot_q + 1):
    share_q = i
    share_score = []
    x += [i]
    for j in range(i):
        same_q = j
        share_score += [sim_score_add_test(tot_q, share_q, same_q, weight)]
    for k in range(i, tot_q + 1):
        share_score += [-1]
    all_score += [share_score]    
    
for a in range(tot_q + 1):
    plt.plot(x, all_score[a])
```
